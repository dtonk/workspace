## Writing a SQL Query
Everything looks good with the data - let's write some SQL queries!

The format for running a query is as follows (it's a little annoying but you'll get used to it):
1. start with `sqldf(
2. three double quotations
3. your sql query
4. three more double quotations
5. end parenthesis 

So, starting with a really simple exmaple, lets SELECT * FROM the dataset we created (called `df`)

```python
# selecting everyting (*) from our df table
sqldf(
    
"""
    SELECT
    *
    FROM df
    LIMIT 5
"""
)
```

Ok, let's get a little more fancy - let's see how many restrooms in each neighborhood

```python
sqldf(
    
"""
    SELECT
        analysis_neighborhood,
        count(uid) restroom_count
    FROM df
    WHERE resource_type = 'restroom'
    GROUP BY analysis_neighborhood
    ORDER BY restroom_count desc
"""
)
```

Neat! But what if we want to export the results in another CSV? 

## Creating a new .csv with your results
Similar to when we created the dataframe, we will need to create a dataframe out of the resuts of the sql query (basically add the data from the SQL query into a table). To do this we run the same query, but make them `=` a table. See below:

```python
#adding the output of a sql query to a dataframe called 'results'
results = sqldf(
    
"""
    SELECT
        analysis_neighborhood,
        count(uid) restroom_count
    FROM df
    WHERE resource_type = 'restroom'
    GROUP BY analysis_neighborhood
    ORDER BY restroom_count desc
"""
)
```

Finally, we will write the data to a new .csv with the `to_csv()` function. 

See more details here: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html

```python
df.to_csv('restrooms_by_neighborhood.csv')
```
